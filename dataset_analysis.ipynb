```python
from statistics import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

df = pd.read_csv("/content/BigML_Dataset_5f50a4cc0d052e40e6000034.csv")
```

```python
#Before

matrix = df.corr()
matrix = matrix[['Power Generated']]
sn.heatmap(matrix, annot=True)
plt.show()
```

```python
df2 = df.copy()
df2['Is Daylight'] = df2['Is Daylight'].where(df2['Is Daylight']==True)
df3 = df2.dropna()
true = mean(df3['Power Generated'])
del df2,df3

df2 = df.copy()
df2['Is Daylight'] = df2['Is Daylight'].where(df2['Is Daylight']==False)
df3 = df2.dropna()
false = mean(df3['Power Generated'])
del df2,df3

x = ["Ishan","False"]
y = [true,false]
plt.bar(x,y,width=0.5)
plt.xlabel("Is Daylight")
plt.ylabel("Avg. Power Generated")
plt.show()
```
